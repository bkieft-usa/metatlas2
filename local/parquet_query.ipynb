# Metatlas2 Targeted Results Parquet Query Tool

Use the .yaml config file to run large-scale queries against all available targeted metabolomics data generated by metatlas2

```python
import os
import logging
from pathlib import Path
import metatlas2.workflows as wfs
import metatlas2.logging_config as lcf

lcf.setup_logging(log_level=logging.INFO, log_to_stdout=True)
logger = lcf.get_logger("parquet_querier")
```

```python
DATA_DIR = Path(os.environ.get('METATLAS_DATA_DIR', Path.home() / '.metatlas2-dev'))
logger.info(f"Set data directory as {DATA_DIR}")
```

```python
wfs.run_parquet_query(
    parquet_path=str(DATA_DIR / "projects/parquet_outputs"),
    params_path=str(DATA_DIR / "configs/parquet_query.yaml")
)
```
